```python
# import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
# check the sample data and shape
df.head()
```

```python
df.shape
```

```python
df.species.value_counts()
```

# Univariate Analysis using only one type of variable

```python
# setosa types
df_setosa=df.loc[df['species']=='setosa']
df_setosa.head()
df_setosa.value_counts().sum()
```

```python
# Versicolor types
df_versicolor=df.loc[df['species']=='versicolor']
df_versicolor.head()
df_versicolor.value_counts().sum()
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_virginica.value_counts().sum()
```

```python
df_setosa.head()
```

```python
# get the plot showing distribution of different species with respect to sepal length
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')

plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')

plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')


plt.xlabel('Sepal length')
plt.show()
```

# Bivariate Analysis

```python
# analysis of species with respect to any of the two features
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,"petal_length",'sepal_width').add_legend();
```
